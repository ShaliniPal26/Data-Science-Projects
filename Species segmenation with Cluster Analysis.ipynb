# Importing relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

# Loading the Data

```python
# Loading the data
data = pd.read_csv('iris_dataset.csv')
# Checking the data
data
```

```python
#creating a scatter plot for the features sepal_length and sepal_width
plt.figure(figsize=(6,4))
sns.scatterplot(data,x='sepal_length',y='sepal_width')
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
```

```python
#creating a copy of data on which we will perform clustering
data_copy = data.copy()

#creating an object of KMeans with 2 cluster
kmeans = KMeans(2)

#fitting the data
kmeans.fit(data_copy)
```

```python
#creating a copy of data again so that we can see the predictions next to our original data
clusters = data.copy()

#adding a column for cluster predictions 
clusters['cluster_prediction']=kmeans.fit_predict(data_copy)
clusters
```

```python
plt.scatter(clusters['sepal_length'], clusters['sepal_width'], c= clusters ['cluster_prediction'], cmap = 'rainbow')
```

```python
from sklearn import preprocessing

scaled_data = preprocessing.scale(data)
scaled_data
```

```python
scaled_kmeans = KMeans(2)

scaled_kmeans.fit(scaled_data)
```

```python
scaled_clusters = data.copy()
scaled_clusters['Scaled_cluster_predictions'] = scaled_kmeans.fit_predict(scaled_data)
```

```python
# create a scatter plot based on two corresponding features (sepal_length and sepal_width; OR petal_length and petal_width)
plt.scatter(scaled_clusters['sepal_length'], scaled_clusters['sepal_width'], c= scaled_clusters['Scaled_cluster_predictions'], cmap = 'rainbow')
```

# Using the Elbow methods - WCSS

```python
wcss = []
# here 10 is a number that keeps track the highest number of clusters we want to use the WCSS method for.
#We have it set at 10 right now, but it is completely arbitrary.

for i in range (1,10):
    kmeans= KMeans(i)
    kmeans.fit(scaled_data)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
wcss
```

```python
no_of_clusters = range(1,10)
plt.plot(no_of_clusters,wcss)
plt.xlabel("Number of Clusters")
plt.ylabel("Within cluster Sum of Squares")
plt.title("The Elbow Method")
plt.show()
```
